```python
def que(board, row, col, n):
    
    for i in range(row):
        if board[i][col] == 1:
            return False

    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True


def queens(board, row, n):

    if row == n:
        solution(board, n)
        return True

    found = False
    for col in range(n):
        if que(board, row, col, n):
            board[row][col] = 1
            found = queens(board, row + 1, n) or found
            board[row][col] = 0  

    return found


def solution(board, n):
    for i in range(n):
        for j in range(n):
            if board[i][j] == 1:
                print("Q", end=" ")
            else:
                print(".", end=" ")
        print()
    print("\n")


# # for example
# n = int(input("Enter number of queens: "))
# board = [[0 for _ in range(n)] for _ in range(n)]

# if not queens(board, 0, n):
#     print("No solution exists")
```
