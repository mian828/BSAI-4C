```python
def water_jug_dfs(capX, capY, goal):
    stack = [((0, 0), [])]   
    
    visited = set()

    while stack:
        (x, y), path = stack.pop()

        if (x, y) in visited:
            continue

        visited.add((x, y))
        print(f"State → X: {x}, Y: {y}")

        # Goal check
        if x == goal or y == goal:
            print("\n Goal Reached!")
            print("Rules Applied:")
            for step in path:
                print("•", step)
            return

        rules = []

        rules.append(((capX, y), "Fill X"))

        rules.append(((x, capY), "Fill Y"))

        rules.append(((0, y), "Empty X"))

        rules.append(((x, 0), "Empty Y"))

        if x + y <= capY:
            rules.append(((0, x + y), "Pour X into Y until X empty"))

        if x + y > capY:
            rules.append(((x - (capY - y), capY), "Pour X into Y until Y filled"))

        if x + y <= capX:
            rules.append(((x + y, 0), "Pour Y into X until Y empty"))

        if x + y > capX:
            rules.append(((capX, y - (capX - x)), "Pour Y into X until X filled"))

        for next_state, rule in rules:
            if next_state not in visited:
                stack.append((next_state, path + [rule]))

    print(" No solution found")

# e.g
capX = int(input("Enter capacity of Jug X: "))
capY = int(input("Enter capacity of Jug Y: "))
goal = int(input("Enter desired output: "))

water_jug_dfs(capX, capY, goal)
```
