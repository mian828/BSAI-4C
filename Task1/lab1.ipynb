```python

import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor


Lab1= r"C:/Users/Riyan/Desktop/PAI LAB/Task1/train.csv"


home_data = pd.read_csv(Lab1)
y = home_data.SalePrice

features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(train_X, train_y)
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("ok: {:,.0f}".format(val_mae))

iowa_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
iowa_model.fit(train_X, train_y)
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation: {:,.0f}".format(val_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=1)

rf_model.fit(train_X,train_y)

predicted_val=rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(val_y,predicted_val)

print("Validation: {}".format(rf_val_mae))
```

```python
test_file_path = r"C:/Users/Riyan/Desktop/PAI LAB/Task1/test.csv"
test_data = pd.read_csv(test_file_path)
```

```python
test_X = test_data[features]
```

```python
test_predictions = iowa_model.predict(test_X)
```

```python
submission = pd.DataFrame({
    "Id": test_data.Id,
    "SalePrice": test_predictions
})

submission.to_csv("submission1.csv", index=False)
```
